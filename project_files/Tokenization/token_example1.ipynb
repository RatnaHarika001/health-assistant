```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

words = ["eating","eaten"]

for word in words:
    print(word+"--->"+stemming.stem(word))
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

paragraph = "TA. P. J. Abdul Kalam, fondly known as the “Missile Man of India,” was a renowned scientist and the 11th President of India who inspired millions through his humility, wisdom, and dedication to the nation. Born in a small town in Tamil Nadu, he rose from modest beginnings to become a key figure in India’s space and defense programs, playing a crucial role in the development of ballistic missiles and launch vehicles. Beyond his scientific achievements, he was deeply passionate about education and youth empowerment, often encouraging students to dream big and work hard to achieve their goals. As President, he was widely loved as the “People’s President” for his simplicity, integrity, and accessibility. His life remains a powerful example of how determination, knowledge, and a strong sense of purpose can transform not only an individual’s destiny but also the future of a nation. 🌟"

sentences = nltk.sent_tokenize(paragraph)

sentences

```

```python
from nltk.corpus import stopwords

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in set(stopwords.words('english'))]
    poss_tag = nltk.pos_tag(words)
    print(poss_tag)
```
